```python
from sklearn.datasets import load_iris
import pandas as pd

i = load_iris()
df = pd.DataFrame(i.data, columns=i.feature_names)
df['target'] = i.target
```

```python
from matplotlib import pyplot as plt
df['sepal length (cm)'].plot(kind = 'hist', bins = 10, title='sepal length (cm)')
plt.show()
# plt.gca().spines[['top','right',]].set_visible(False)
```

```python
plt.figure(figsize=(8,6))
plt.boxplot([
    df[df['target'] == 0]['sepal length (cm)'],
    df[df['target'] == 1]['sepal length (cm)'],
    df[df['target'] == 2]['sepal length (cm)']
],
labels= ['Target 0', 'Target 1', 'Target 2']
)
plt.xlabel('Target Class')
plt.ylabel('Sepal Length (cm)')
plt.title('Distribution of sepal length by target class')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'], cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
_ = plt.colorbar(label='Target Class')
plt.show()
```

```python
import seaborn as sns
species_counts = df['target'].value_counts()

# create a pie chart
plt.figure(figsize=(8,5))
sns.set_palette('pastel')
explode = [0.1, 0.1, 0.1]
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Proportion of Iris Species')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# create a scatter plot matrix
sns.pairplot(df, hue='target', palette='viridis')
plt.title('Iris dataset scatter plot matrix')
plt.show()
```
